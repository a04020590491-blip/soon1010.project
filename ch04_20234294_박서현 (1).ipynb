```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/p/ch04/winter.jpg') # Check and correct this file path

if img is not None:
    print(img.shape)
    glmg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print(glmg1.shape)
    plt.imshow(glmg1, cmap='gray')
else:
    print("이미지 로드 실패. 파일 경로를 확인하세요.")
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
    return val
```

```python
gPoint = np.zeros((glmg1.shape[0], glmg1.shape[1]), dtype=np.ubyte)
```

```python
# Define the threshold value (adjust as needed)
th = 80 # Example threshold value

for h in range(glmg1.shape[0]):
  for w in range(glmg1.shape[1]):
    if glmg1[h,w] > th:
      gPoint[h,w] = 255
    else: # Corrected indentation
      gPoint[h,w] = 0

plt.imshow(gPoint, cmap='gray')
```

```python
# Define the threshold value (adjust as needed)
th = 120 # Example threshold value

for h in range(glmg1.shape[0]):
  for w in range(glmg1.shape[1]):
    if glmg1[h,w] > th:
      gPoint[h,w] = 255
    else: # Corrected indentation
      gPoint[h,w] = 0

plt.imshow(gPoint, cmap='gray')
```

```python
numOfClip = 3
modval = 256/numOfClip

for h in range(glmg1.shape[0]):
  for w in range(glmg1.shape[1]):
    pointval = np.int32(glmg1[h,w]) % modval
    gPoint[h,w] = pointval

plt.imshow(gPoint, cmap='gray', vmin=0, vmax=255)
```
